```python
# Idea from https://medium.com/@praveen.orvakanti/this-will-help-you-score-95-percentile-in-the-kaggle-titanic-ml-competition-aa2b3fd1b79b

df['Relatives'] = df['SibSp'] + df['Parch']
```

```python
from sklearn.preprocessing import StandardScaler

data = df[['Pclass','SibSp','Parch', 'Relatives', 'Fare']]
# define min max scaler
scaler = StandardScaler()
# transform data
scaled = scaler.fit_transform(data)

df[['Pclass','SibSp','Parch', 'Relatives', 'Fare']] = scaled

df
```

```python
df.isna().sum()
```

```python
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression

df.Sex = pd.get_dummies(df['Sex'],drop_first=True)

X_train = df.loc[df['Age'].notna()][['Pclass', 'Sex', 'SibSp', 'Parch', 'Relatives']]
Y_train = df.loc[df['Age'].notna()]['Age']
X_test = df.loc[df['Age'].isna()][['Pclass', 'Sex', 'SibSp', 'Parch', 'Relatives']]

# Source: https://www.kaggle.com/pratsiuk/valueerror-unknown-label-type-continuous/notebook
lab_enc = preprocessing.LabelEncoder()
Y_train = lab_enc.fit_transform(Y_train)

logmodel = LogisticRegression()
logmodel.fit(X_train,Y_train)

predictions = logmodel.predict(X_test)

predictions
```

```python
df.Age[df.Age.isna()] = predictions
df
```

```python
data = pd.DataFrame(df['Age'])
# define min max scaler
scaler = StandardScaler()
# transform data
scaled = scaler.fit_transform(data)

df['Age'] = scaled

df
```

```python
df.drop(columns=['Name','Cabin','Ticket','PassengerId'], inplace=True)

sex = pd.get_dummies(df['Sex'],drop_first=True)
embarked = pd.get_dummies(df['Embarked'],drop_first=True)
df = pd.concat([df,sex,embarked],axis=1)
df.drop(['Sex','Embarked'],axis=1,inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report


X= df.drop('Survived',axis=1)
y= df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)

predictions = logmodel.predict(X_test)

print('Train', logmodel.score(X_train, y_train)*100)
print('Test', logmodel.score(X_test, y_test)*100)
```
